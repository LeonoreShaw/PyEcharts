```python
from pyecharts.charts import Bar
from pyecharts import options as opts

from pyecharts.globals import ThemeType, RenderType
from pyecharts.faker import Faker
```

```python
bar = Bar(
    # 初始化配置项，可选项
    init_opts=opts.InitOpts(
        renderer=RenderType.SVG, # 渲染模式，可选'render_type=RenderType.CANVAS'或'render_type=RenderType.SVG'
        page_title="Bar-RenderType.SVG", # 网页标题
        theme=ThemeType.LIGHT, # 主题
        bg_color="transparent", # 背景颜色
        
        
    )
)

bar.add_xaxis(["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"])
bar.add_yaxis("商家A", [5, 20, 36, 10, 75, 90])
bar.add_yaxis("商家B", [15, 6, 45, 20, 35, 66])

# 全局配置项，可选项
bar.set_global_opts(
    title_opts=opts.TitleOpts(
        title="主标题", 
        title_link="https://www.baidu.com", # 主标题超链接
        subtitle="副标题",
        subtitle_link="https://www.baidu.com", # 副标题超链接
        )
    
)


bar .render_notebook()

```

```python
bar1 = Bar(
    init_opts=opts.InitOpts(
        renderer=RenderType.CANVAS, # 渲染模式，可选'render_type=RenderType.CANVAS'或'render_type=RenderType.SVG'
        page_title="Bar-RenderType.html", # 网页标题
        theme=ThemeType.LIGHT, # 主题
        bg_color="transparent", # 背景颜色
        
        
    )
)
x = Faker.choose()
bar1.add_xaxis(x)
bar1.add_yaxis("A", Faker.values())
bar1.add_yaxis("B", Faker.values())


# 全局配置项，可选项
bar1.set_global_opts(
    title_opts=opts.TitleOpts(
        title="主标题", 
        title_link="https://www.baidu.com", # 主标题超链接
        subtitle="副标题",
        subtitle_link="https://www.baidu.com", # 副标题超链接
        
        # 标题位置
        pos_left="5px", # 标题水平位置，可选'left', 'center', 'right'
        ),
    # 缩放配置项
    datazoom_opts=opts.DataZoomOpts(
        is_show=True, # 是否显示
        type_="slider", # 缩放类型，可选'slider', 'inside'
        is_realtime=True, # 是否实时显示
    ),
    # 图例配置项
    legend_opts=opts.LegendOpts(
        is_show=True, # 是否显示
        type_="scroll", # 图例类型，可选'plain', 'scroll'
        pos_right="5px", # 图例水平位置，可选'left', 'center', 'right'
        orient="horizontal", # 图例垂直位置，可选'horizontal', 'vertical'
        legend_icon="circle", # 图例形状，可选'circle', 'rect', 'roundRect', 'triangle', 'diamond', 'pin', 'arrow', 'none'
    ),
    # VisualMap配置项
    visualmap_opts=opts.VisualMapOpts(
        is_show=True, # 是否显示
        type_="color", # 映射类型，可选'color', 'size'
        
    ),
    # 提示框配置项
    tooltip_opts=opts.TooltipOpts(
        is_show=True, # 是否显示
    ),
    # 坐标轴配置项
    xaxis_opts=opts.AxisOpts(
        
    ),
    
)

bar1 .render(x[0] + ".html")
```
