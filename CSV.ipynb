```python
import pandas as pd
from pyecharts.charts import Line
from pyecharts import options as opts

# Step 1: Read data from CSV file
df = pd.read_csv('annual-balance-sheets-2007-2021-provisional.csv')  # Replace 'data.csv' with your CSV file name
df
```

```python

y_data = df['Values']  # Replace 'y_column' with the actual column name for y-axis


xdata = list(range(6556))  # Generate a list from 0 to 100
xdf = pd.DataFrame(xdata, columns=['x'])
print(type(xdf))
xdf
```

```python
xd = xdf['x']  # Replace 'x_column' with the actual column name for x-axis
xdf
```

```python
y_data
```

```python

# Step 2: Create a Line chart
line_chart = Line()
# xd的前100个数据
xxd = xd[:100]
# y_data的前100个数据
yy_data = y_data[:100]

# line_chart.add_xaxis(xdf['x'])
# line_chart.add_yaxis("Y Data", y_data)

line_chart.add_xaxis(xxd)
line_chart.add_yaxis("Data", yy_data)


# Step 3: Set chart title and axis labels
line_chart.set_global_opts(
    title_opts=opts.TitleOpts(title="Line Chart", pos_left='center', pos_top='top', subtitle='subtitle', title_textstyle_opts=opts.TextStyleOpts(color='red', font_size=20, font_style='italic', font_weight='bold')),
    xaxis_opts=opts.AxisOpts(name="X"),
    yaxis_opts=opts.AxisOpts(name="Y"),
    legend_opts=opts.LegendOpts(pos_right='right', pos_top='top', orient='vertical', textstyle_opts=opts.TextStyleOpts(font_size=20, font_style='italic', font_weight='bold')),
    datazoom_opts=opts.DataZoomOpts(
        is_show=True, # 是否显示
        type_="inside", # 缩放类型，可选'slider', 'inside'
        
    ),
)
# Step 4: Hide data values
line_chart.set_series_opts(label_opts=opts.LabelOpts(is_show=False))



# Step 4: Render the chart to an HTML file
# line_chart.render("line_chart.html")
line_chart.render_notebook()
```
