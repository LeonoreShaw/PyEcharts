```python
import pandas as pd

from pyecharts.charts import Bar
from pyecharts.charts import Line

from pyecharts import options as opts

from pyecharts.globals import ThemeType, RenderType
from pyecharts.faker import Faker
```

```python
# 读取数据
df = pd.read_csv('Business-employment-data-september-2022-quarter-csv.csv', encoding='utf-8')
df
```

```python
type(df['Period'])
```

```python
line = (
    Line()
    .add_xaxis(pd.Series(range(0, len(df['Period']))))
    .add_yaxis('Data', df['Data_value'])
)
line.set_global_opts(
    datazoom_opts=opts.DataZoomOpts(
        is_show=True, # 是否显示
        type_="slider", # 缩放类型，可选'slider', 'inside'
        is_realtime=True, # 是否实时显示
    ),
)
line.set_series_opts(label_opts=opts.LabelOpts(is_show=False))
line.render_notebook()
```
